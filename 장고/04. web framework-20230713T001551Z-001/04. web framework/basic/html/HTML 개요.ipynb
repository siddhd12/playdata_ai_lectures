### [Outline Algorithm](https://seulbinim.github.io/WSA/html-basic.html#html4-01-xhtml1-0%EA%B3%BC-html5%EC%9D%98-%EC%B0%A8%EC%9D%B4%EC%A0%90)
HTML5에서는 정보 구조를 명확히 할 수 있도록 `아웃라인 알고리즘(Outline Algorithm)`이라는 개념이 도입되었습니다. 아웃라인 알고리즘은 웹 페이지의 정보 구조를 판별할 수 있는 개념으로 책의 목차와 비슷합니다.

아래 HTML문서를 [HTML5 Ouline 판별 웹 사이트](https://gsnedders.html5.org/outliner/)에 적용하면, 아래와 같은 결과를 얻을 수 있습니다.

1. HTML5
2. 메인 메뉴
3. 콘텐츠 모델 이란?
4. 콘텐츠 모델의 분류
5. 메타데이터 콘텐츠
6. 플로우 콘텐츠

```html
<!-- 아웃라인이 적용되지 않은 문서 -->

<!DOCTYPE html>
<html lang="ko-KR">
<head>
  <meta charset="utf-8">
  <title>div 요소를 사용한 문서의 아웃라인</title>
</head>
<body>
  <div id="wrapper">
    <div id="header">
      <div class="branding">
        <h1>HTML5</h1>
      </div>
      <div class="gnb">
        <h1>메인 메뉴</h1>
        <ul>
          <li><a href="#">콘텐츠 모델</a></li>
          <li><a href="#">아웃라인 알고리즘</a></li>
        </ul>
      </div>
    </div>
    <div id="main">
      <div class="section">
        <h1>콘텐츠 모델 이란?</h1>
        <p>어떤 요소에 어떤 콘텐츠를 포함해야 하는지 또 어떤 요소가 어떤 요소를 포함할 수 
         있는지를 정의한 것을 말합니다.</p>
      </div>
      <div class="section">
        <h1>콘텐츠 모델의 분류</h1>
        <div class"article">
          <h1>메타데이터 콘텐츠</h1>
          <p>Metadata 콘텐츠는 다른 콘텐츠의 표현과 행동을 설정하거나 문서와 문서의 
           관계를 설정할 때 사용합니다.</p>
        </div>
        <div class="article">
          <h1>플로우 콘텐츠</h1>
          <p>Flow 콘텐츠는 웹 페이지상에 직접 표시되지 않는 일부 메타데이터를 
           제외하고 거의 모든 요소들이 이 그룹에 속합니다.</p>
        </div>
      </div>
    </div>
    <div id="footer">
      <p>Copyright &copy; webCafe</p>
    </div>
  </div>
</body>
</html>
```

아래 HTML문서를 [HTML5 Ouline 판별 웹 사이트](https://gsnedders.html5.org/outliner/)에 적용하면, 아래와 같은 결과를 얻을 수 있습니다.

이렇게 section이나 article 등을 적절하게 이용하면, 문서의 정보뿐만 아니라 문서의 구조(목차)도 전달할 수 있습니다. 

```html
<!-- 아웃라인이 적용된 문서 -->

<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>섹셔닝 요소를 사용한 문서의 아웃라인</title>
</head>
<body>
 <div id="wrapper">
  <header id="header">
   <div class="branding">
    <h1>HTML5</h1>
   </div>
   <nav class="gnb">
    <h1>메인 메뉴</h1>
    <ul>
     <li><a href="#">콘텐츠 모델</a></li>
     <li><a href="#">아웃라인 알고리즘</a></li>
    </ul>
   </nav>
  </header>
  <div id="main">
   <section class="section">
    <h1>콘텐츠 모델 이란?</h1>
    <p>어떤 요소에 어떤 콘텐츠를 포함해야 하는지 또 어떤 요소가 어떤 요소를 포함할 수 
    있는지를 정의한 것을 말합니다.</p>
   </section>
   <section class="section">
    <h1>콘텐츠 모델의 분류</h1>
    <article class"article">
     <h1>메타데이터 콘텐츠</h1>
     <p>Metadata 콘텐츠는 다른 콘텐츠의 표현과 행동을 설정하거나 문서와 문서의 
     관계를 설정할 때 사용합니다.</p>
    </article>
    <article class="article">
     <h1>플로우 콘텐츠</h1>
     <p>Flow 콘텐츠는 웹 페이지상에 직접 표시되지 않는 일부 메타데이터를 
     제외하고 거의 모든 요소들이 이 그룹에 속합니다.</p>
    </article>
   </section>
  </div>
  <footer id="footer">
   <p>Copyright &copy; webCafe</p>
  </footer>
 </div>
</body>
</html>
```

### 유효성 검사(Markup Validation)
W3C에서 제공하는 [유효성 검사(Markup Validation)](https://validator.w3.org/) 웹 페이지에 html 파일을 올리면 자동으로 HTML5 문서의 유효성 여부를 검사할 수 있습니다.  

```html
<!DOCTYPE html>
<html lang="en">
<head>
 <meta charset="utf-8">
 <title>HTML5 validation Error</title>
</head>
<body>
 <strong>
  <a href="http://validator.w3.org/">
   <section>W3C Markup Validation</section>
  </a>
 </strong>
</body>
</html>
```

위의 문서를 이용하여 유효성 검사를 하면 아래와 같은 오류 메세지를 볼 수 있습니다.  

오류 메세지의 내용은 `<strong>` 요소의 자손으로 `<section>` 요소를 포함하여 발생한 오류입니다.

image.png

아래 문서로 수정하여 다시 유효성 검사를 진행하면 통과했다는 결과를 받을 수 있습니다.

```html
<!DOCTYPE html>
<html lang="en">
<head>
 <meta charset="utf-8">
 <title>HTML5 validation Error</title>
</head>
<body>
 <section>
  <a href="http://validator.w3.org/">
   <strong>W3C Markup Validation</strong>
  </a>
 </section>
</body>
</html>
```

## [메타데이터 관련 요소](https://seulbinim.github.io/WSA/metadata.html)

### meta 요소

- 문자 인코딩 지정

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <!-- HTML5에 추가된 방식 -->
 <meta charset="utf-8">
 <!-- 기존 방식 -->
 <meta http-equiv="content-type" content="text/html; charset=euc-kr">
 <title>문자 인코딩 지정하기</title>
</head>
<body>
 <h1>문자 인코딩</h1>
 <p>문자 인코딩 지정은 HTML5 방식과 기존 방식 모두 허용됩니다.</p>
</body>
</html>
```

- 키워드 지정 

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>검색 키워드 지정</title>
 <meta name="keywords" content="웹접근성,웹표준,HTML,웹콘텐츠 접근성 지침">
</head>
<body>
 <h1>검색 키워드</h1>
 <p>meta 요소를 이용하면 웹 페이지 홍보의 수단으로 검색 키워드를 지정할 수 있습니다. 웹 페이지에 검색 키
  워드를 지정하면 검색할 때 웹 사이트가 상위에 노출되는 효과를 얻을 수 있습니다. meta 요소를 이용하여 검
  색 키워드를 지정할 때는 검색 키워드를 콤마(,)로 구분하여 선언합니다.</p>
</body>
</html>
```


- 다양한 문서 정보 지정 

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>문서 정보 삽입하기</title>
 <meta name="keywords" content="웹 접근성, 웹 표준, 품질마크, 웹 콘텐츠 접근성 지침">
 <meta name="subject" content="웹 접근성 전문가 소개">
 <meta name="description" content="웹 접근성에 대한 개요와 웹 접근성 품질마크 소개 및 
 한국형 인터넷 웹 콘텐츠 지침에 대한 내용입니다.">
 <meta name="author" content="seulbinim@gmail.com(Teresa Kim), 김데레사">
 <meta name="robots" content="index,follow">
 <meta name="copyright" content="copyrights 2010 WAP corp.">
 </head>
<body>
 <h1>다양한 문서 정보 지정</h1>
 <p>화면에 보이는 콘텐츠는 아니지만 웹 페이지의 다양한 문서 정보를 지정할 때도 meta 요소를 사용할 수 있습니다. 문자 코드 세트와 키워드 외에도 subject를 이용하여 문서 제목 정보를 제공하고, description값으로 웹 페이지의 요약 정보, 제작자 정보, 저작권 정보 및 검색 로봇 제어 여부 등의 문서 정보를 지정할 수 있습니다.</p>
</body>
</html>
```

- [반응형 설정을 위한 viewport 설정](https://developer.mozilla.org/en-US/docs/Web/HTML/Viewport_meta_tag) 

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>뷰포트 지정하기</title>
 <meta name="viewport" content="width=device-width, initial-scale=1.0">
</head>
<body>
 <h1>뷰포트 지정하기</h1>
 <p>뷰포트는 화면 상에 보이는 영역을 의미하는데 처음 모바일 사파리(Safari)의 경우 
  meta 요소에서 viewport를 지정하여 화면 크기를 조정할 수 있게 제공 하였습니다. 
  현재 대다수의 웹브라우저 들이 이를 지원하고 있습니다.</p>
</body>
</html>
```

### title 요소

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>웹 문서의 제목</title>
</head>
<body>
 <h1>웹 문서의 제목</h1>
 <p>웹 문서의 제목을 선언하려고 할 때 사용하는 요소로, 문서마다 유일한 내용으로 구성해야 합니다. 
  시각 장애인의 경우에는 화면 낭독기를 이용하여 콘텐츠를 탐색합니다. 
  따라서 title 요소는 문서를 구별할 수 있는 첫 번째 관문이라고 할 수 있습니다.</p>
</body>
</html>
```

### section 
- section 요소는 일반적으로 콘텐츠를 장이나 절 등으로 구분하고자 할 경우에 사용합니다. 
- 장이나 절 등으로 구분되기 때문에 section 요소 안에는 헤딩 관련 요소를 사용하여 해당 섹션의 제목을 지정해야 합니다. 
- section 요소에 제목이 없는 경우에는 장이나 절의 섹션이라고 할 수 없기 때문에 반드시 제목을 제공하도록 주의를 기울여야 합니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>section 요소</title>
</head>
<body>
 <section>
  <h1>섹션(Section) 요소</h1>
  <section>
   <h2>섹션(Section) 요소의 사용</h2>
   <p>section 요소는 일반적으로 콘텐츠를 장이나 절 등으로 구분하고자 할 경우나 탭 형식으로 되어
    있는 콘텐츠를 마크업할 때 사용합니다.</p>
   <p>장이나 절 등으로 구분되기 때문에 반드시 section 요소 안에 제목을 의미하는 헤딩 관련 요소
    가 포함되어야 합니다.</p>
  </section>
  <section>
   <h2>섹션 요소의 콘텐츠 모델</h2>
   <p>section 요소는 섹셔닝 콘텐츠에 속한 요소로 아웃라인에 영향을 줍니다.</p>
  </section>
 </section>
</body>
</html>
```

### article
- article 요소는 플로우 콘텐츠(Flow Content)와 섹셔닝 콘텐츠(Sectioning Content)에 해당하는 요소로, 독립된 하나의 콘텐츠를 의미합니다. 그 예로는 블로그의 포스트나 뉴스 본문 등을 들 수 있습니다. 
- 독립된 하나의 콘텐츠로 분류할 때 해당 콘텐츠를 별도로 배포하거나 재사용하기 위한 경우에 article 요소를 사용하면 됩니다.
- article 요소 안에 article 요소를 포함하여 마크업할 수 있으며, 이 경우 article 요소 안에 포함된 자손 article 요소는 부모 article 요소와 직접적으로 관련 있는 콘텐츠여야 합니다. 주로 블로그 포스트의 답글 등이 이에 해당합니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>article 요소</title>
</head>
<body>
 <article class="post">
  <h1>책 그리고 커피와 함께한 생일</h1>
  <p>느즈막이 일어난 생일 아침. 요가로 몸을 풀어준 뒤 책 몇 권을 가방에 넣고 가까운 커피 전문점으로 향했
   다. 커피는 아메리카노에 샷 추가로 주문하고 햇살이 따뜻하게 비치는 창가에 자리를 잡고 가지고 온 책을 펼쳤
   다. </p>
  ……
  <footer>
   <p>슬비</p>
   <time datetime="2013-10-04T11:30:00+09:00">10월 4일</time>
  </footer>
  <article class="reply">
   <p>생일이셨군요. 늦었지만 축하 드려요. 커피한잔과 책을 읽으며 여유롭고 한가하게 보내신 모습이 부러
    운데요. 담 기회에는 함께 커피 마시며 얘기 나눌 수 있으면 좋겠네요. </p>
   <footer>
    <p>미희</p>
    <time datetime="2013-10-05T13:45:30+09:00">">10월 5일</time>
   </footer>
  </article>
 </article>
 ……
</body>
</html>
```

### nav
- nav 요소는 내비게이션을 마크업할 때 사용합니다. 
- 내비게이션은 메인 메뉴나 서브 메뉴 등을 의미하며, 서로 다른 페이지로 이동할 수 있는 링크 항목을 포함합니다. 
- 그러나 모든 목록 형태의 하이퍼링크 콘텐츠에 nav 요소를 사용하는 것은 바람직하지 않습니다.

nav 요소는 페이지의 주요 내비게이션에 사용해야 한다고 규정하고 있습니다. 주요 내비게이션에 포함되지 않는 대표적인 콘텐츠로는 사이트 이용 안내 등의 footer 링크 목록을 들 수 있습니다. 이러한 링크 목록 콘텐츠의 경우에는 사이트의 주요 내비게이션이라 보기 어려우므로 nav 요소를 사용하지 않는 것이 적절합니다.


```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>nav 요소</title>
</head>
<body>
 <header>
  <h1>
   <a href="index.html"><img src="images/logo.gif" alt="WebCafe"></a>
  </h1>
  <nav>
   <ul>
    <li><a href="standard.html">웹표준</a></li>
    <li><a href="accessibility.html">웹 접근성</a></li>
    <li><a href="qna.html">묻고 답하기</a></li>
    <li><a href="pds.html">자료실</a></li>
    <li><a href="join.html">회원가입</a></li>
    <li><a href="sitemap.html">사이트맵</a></li>
   </ul>
  </nav>
 </header>
 ...
</body>
</html>
```

### aside
- aside 요소는 플로우 콘텐츠(Flow Content)와 섹셔닝 콘텐츠(Sectioning Content)에 해당하는 요소로, 메인 콘텐츠와 관련이 적은 콘텐츠를 마크업할 때 사용할 수 있습니다. 
- 예를 들어 사이드 바, 광고 등의 콘텐츠가 aside 요소로 마크업하기 적절한 콘텐츠라고 할 수 있습니다. 
- 사이드 바와 광고 등의 콘텐츠는 삭제되더라도 메인 콘텐츠에 영향을 끼칠 수 있는 콘텐츠가 아닌 부가적인 콘텐츠이기 때문입니다.

그러나 aside 요소에 대한 사용은 명확히 정의하기 어려운 경우도 많습니다. 가령 제작자의 의도에 따라 블로그의 답글을 section 요소나 article 요소가 아닌 aside 요소를 사용하여 본문 콘텐츠와 직접 연관시키지 않을 수도 있기 때문입니다. aside 요소는 이렇게 명확하지 않은 부분 때문에 사용하기가 꺼려질 수 있습니다. aside 요소는 콘텐츠의 틀을 정해놓고 사용하기보다 콘텐츠의 중요성이나 의미를 고려하여 사용하기 바랍니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>aside 요소</title>
</head>
<body>
 <section>
  ...
 </section>
 <aside>
  <h1>국내 추천 여행지 Best5</h1>
  <ul>
   <li><a href="http://www.boseong.go.kr/ko">보성 녹차밭</a></li>
   <li><a href="http://www.samyangranch.co.kr">평창 대관령 목장</a></li>
   <li><a href="http://naejang.knps.or.kr">정읍 내장산 국립공원</a></li>
   <li><a href="http://www.naesosa.org">부안 내소사</a></li>
   <li><a href="http://www.oedebotania.com">거제 외도 해상공원</a></li>
  </ul>
 </aside>
 ...
</body>
</html>
```

### header
- header 요소는 페이지 헤더나 섹션 헤더로 사용할 수 있으며, 페이지 헤더의 경우 사이트 제목, 로고, 검색 영역 등을 콘텐츠로 포함할 수 있습니다. 
- 또한 section 요소 내에서 사용할 경우, 해당 섹션의 헤더의 의미를 가지게 됩니다. 이때 섹션 헤더의 경우 header 요소에 제목을 포함하는 것은 필수가 아닙니다.

다음은 header 요소의 사용 예입니다. 섹션의 차례나 검색 폼, 관련 로고 등을 감싸는 용도로도 사용할 수 있습니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>header 요소</title>
</head>
<body>
 <header> <!-- 본문 헤더 -->
  <a href="index.html">
   <img src="images/logo.gif" alt="WebCafe"></a>
  …
 </header>
 <section>
  <header> <!-- 섹션 헤더 -->
   …
  </header>
  …
 </section>
</body>
</html>
```

### footer
- footer 요소는 헤더와 마찬가지로 페이지 푸터와 섹션 푸터로 구분할 수 있으며, 페이지 푸터의 경우 사이트의 연락처 정보 및 저작권 등을 포함할 수 있습니다. 
- 섹션 푸터로 사용하였다면 해당 섹션의 푸터 정보를 의미합니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>footer 요소</title>
</head>
<body>
 <section>
  …
  <footer> <!-- 섹션 푸터 -->
   …
  </footer>
 </section>
 <footer> <!-- 본문 푸터 -->
  <address>서울시 마포구 상수동 324-1 한주빌딩 5층 | seulbinim@gmail.com
  </address>
  <p class="copyright">
   <small>Copyright &copy; Web Cafe ALL RIGHTS RESERVED.</small>
  </p>
 </footer>
</body>
</html>
```

## [그룹화 관련 요소](https://seulbinim.github.io/WSA/grouping.html#div-%EC%9A%94%EC%86%8C)

### div
- div 요소는 의미를 가지고 있지 않지만 디자인상의 이슈나 자바스크립트로 특정 콘텐츠 블록을 컨트롤하는 데 사용할 수 있습니다. 
- 만약 콘텐츠 블록이 섹션의 의미를 가지고 있다면 div 요소를 사용하는 것은 적절하지 않습니다.

```html
<html lang="ko-KR">
<head>
 <meta charset="UTF-8">
 <title>div 요소</title>
</head>
<body>
 <section class="coffee">
  <h1>커피(Coffee)</h1>
  <div class="box">
   <p><em>아메리카노(Americano)</em>는 에스프레소에 뜨거운 물을 더하여 
    마시는 방식으로 그 농도는 일반적인 드립 커피와 비슷하지만 풍미는 다릅니다. 
    아메리카노의 농도는 에스프레소의 '샷' 수와, 더해지는 물의 양에 따라 달라집니다.</p>
   <p><em>카페라떼(Cafe Latte)</em>는 커피와 우유라는 뜻을 지닌 이탈리아어로서 
    커피와 뜨거운 우유를 곁들여 마시는 커피의 일종입니다. 카페라테는 전 세계에서 
    찾아볼 수 있는 커피의 종류중 하나로서 카푸치노와 에스프레소와 함께 가장 흔한 종류 중 
    하나이기도 합니다.</p>
   <p><em>카푸치노(Cappuccino)</em>는 에스프레소, 뜨거운 우유, 그리고 우유 거품을 
    재료로 만드는 이탈리아의 커피 음료로 카페라테에 비해 우유의 양이 훨씬 적습니다. 
    카푸치노는 전통적으로 자기로 된 컵에 서빙되는데, 이는 자기가 유리나 종이보다 열을 
    더 잘 보존하기 때문입니다.</p>
  </div>
 </section>
</body>
</html>
```

### main
- main 요소는 HTML5에 새롭게 추가되었으며, body 영역의 주요 콘텐츠 블록을 그룹화할 때 사용합니다. 
- 그러나 main 요소로 사이드 바나 검색 등의 콘텐츠 영역을 그룹화해서는 안 됩니다. 
- 또한 main 요소는 body 요소 내에서 한 번만 사용해야 합니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>main 요소</title>
</head>
<body>
 <main>
  <section>
   <h1>아메리카노(Americano)</h1>
   <p><em>아메리카노</em>는 에스프레소에 뜨거운 물을 더하여 먹는 방식으로 그 농도는 일반적인 드
    립 커피와 비슷하지만 풍미는 다릅니다. 아메리카노의 농도는 에스프레소의 '샷' 수와, 더해지는 물의 양
    에 따라 달라집니다.</p>
  </section>
  <section>
   <h1>카페라떼(Cafe Latte)</h1>
   <p><em>카페라떼</em>는 커피와 우유라는 뜻을 지닌 이탈리아어로서 커피와 뜨거운 우유를 곁들여
    마시는 커피의 일종입니다. 카페라테는 전 세계에서 찾아볼 수 있는 커피의 종류중 하나로서 카푸치노와 에
    스프레소와 함께 가장 흔한 종류 중 하나이기도 합니다.</p>
  </section>
 </main>
 …
</body>
</html>
```

### p
- 텍스트를 단락으로 정의할 때 사용합니다. 
- 단락 요소 안에는 b, img 등 인라인 요소와 텍스트만 포함할 수 있으며, 블록 요소는 사용할 수 없습니다. 
- 단락 안에서 강제로 줄바꿈해야 하는 경우에는 ‘line break’의 의미인 br 요소를 사용합니다. 
- br 요소는 빈 요소(empty element)이며 줄바꿈을 위해 사용할 수 있지만, 시각적 효과 등을 위해 남용하지 않아야 합니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
<meta charset="utf-8">
<title>단락 요소</title>
</head>
<body>
<h1>단락 요소</h1>
<p>텍스트를 단락으로 정의할 때 사용합니다. 단락 요소 안에는 &lt;b&gt;, &lt;img&gt;등 <strong>인라인 요소</strong>와 <strong>텍스트만 포함</strong>할 수 있으며 블록 요소는 사용할 수 없습니다.</p>
<p>제목 요소 및 단락 요소는 각 콘텐츠 위/아래 방향으로 <strong>마진</strong>이 적용되어 있어 <strong>여백</strong>이 생깁니다. 해당 여백으로 이동 시각적으로 콘텐츠 영역이 구분되어 있음을 알 수 있습니다.</p>
</body>
</html>
```

### address
- address 요소는 주소 등 연락처 정보를 마크업할 때 사용합니다. 
- HTML5에서는 address 요소의 사용을 명확하게 제시하고 있는데, address 요소는 해당 페이지 또는 해당 포스트의 관리자 및 저자의 연락처 정보를 의미합니다. 
- 모든 주소 콘텐츠를 address 요소를 사용하여 마크업해서는 안 됩니다.

```html
<!DOCTYPE html>
<html lang="ko-KR">
<head>
 <meta charset="utf-8">
 <title>address 요소</title>
</head>
<body>
 …
 <footer>
  <address>
   서울시 마포구 상수동 324-1 한주빌딩 5층 | seulbinim@gmail.com
  </address>
  …
 </footer>
</body>
</html>
```

### 비순서형 목록 요소(ul, li)
- ul 요소는 비순서형 목록을 의미합니다. 
- ul 요소에 올 수 있는 자식 요소는 li 요소만 가능하며, li 요소는 목록 항목을 의미합니다. 
- ul 요소는 목록 항목의 순서를 변경하더라도 의미가 변하지 않는 콘텐츠를 마크업할 때 사용합니다.

```html
<html lang="ko">
<head>
<meta charset="utf-8">
<title>ul, li 요소</title>
</head>
<body>
<section>
<h1>커리(Curry) 재료</h1>
<ul>
<li>카레가루 100그램</li>
<li>닭가슴살 200그램</li>
<li>감자 1개</li>
<li>당근 1/2개</li>
<li>양파 1/2개</li>
</ul>
</section>
</body>
</html>
```

### 순서형 목록 요소(ol, li)
- ol 요소는 순서형 목록을 의미합니다. 
- ol 요소에 올 수 있는 자식 요소는 ul 요소와 마찬가지로 li 요소만 가능합니다. 
- ol 요소는 목록 항목의 순서가 의미상 중요할 경우에 사용합니다. 
- ol 요소는 HTML5에서 reversed 속성이 추가되어 목록의 렌더링 순서를 오름차순 또는 내림차순으로 지정할 수 있습니다.

```html
<!DOCTYPE html>
<html lang="ko">
<head>
<meta charset="utf-8">
<title>ol, li 요소</title>
</head>
<body>
<section>
<h1>인기 검색어 순위</h1>
<ol>
<li>HTML5</li>
<li>CSS3</li>
<li>JQuery</li>
<li>웹접근성</li>
<li>웹표준</li>
</ol>
<ol reversed>
<li>HTML5</li>
<li>CSS3</li>
<li>JQuery</li>
<li>웹접근성</li>
<li>웹표준</li>
</ol>
</section>
</body>
</html>
```

### 정의형 목록 요소(dl, dt, dd)
- dl은‘Definition List’의 약자로, 용어 정의 리스트를 생성할 때 사용합니다. 최근에는 웹 표준의 적용으로 의미 있는 콘텐츠에 대해 관심이 많아지고 있지만, 디테일한 의미 부여가 XML처럼 자유롭지 못하기 때문에 제한된 요소 중에서 적당한 요소로 마크업하기 위한 고민이 깊어지고 있습니다.

- 그 중에서도 dl 요소를 이용한 콘텐츠 마크업이 주목을 받고 있습니다. 왜냐하면 dl 요소의 경우 단순히 정의형 목록(용어 제목, 용어 설명 등)만이 아니라 주종 관계가 성립되는 콘텐츠에 광범위하게 사용할 수 있기 때문입니다. dl 요소의 사용 형식은 다음과 같습니다.

- dl 요소는 이름과 값이 쌍으로 구성되어 있는 콘텐츠인 경우에 사용할 수 있으며, 용어와 그에 대한 설명을 마크업할 때도 적절하게 사용할 수 있습니다. 이때 용어의 이름은 dt 요소로, 용어에 대한 설명은 dd 요소로 마크업하며 dt 요소와 dd 요소는 1 : 1, 1 : n(여러 개), n(여러 개) : 1의 경우에도 사용할 수 있습니다.

```html
<!DOCTYPE html>
<html lang="ko">
<head>
<meta charset="utf-8">
<title>dl, dt, dd 요소</title>
</head>
<body>
<section>
<dl>
<dt>XHTML</dt>
<dd>HTML은 Hyper Text Markup Language의 약자로, 웹 문서를 제작하기 위한 언어로 설계되었습니다. 그래서 범용적 전자 문서나 전자데이터를 표현하는 기능은 충분하지 않았습니다. 이러한 요구로 인해 등장한 것이 바로 XML(eXtensible Makeup Language)이며, XHTML은 HTML을 XML로 재구축한 마크업 언어입니다. 그러므로 HTML과 XHTML은 동일한 요소를 가지며 두 가지 마크업 언어 모두 웹 문서를 제작할 때 사용되는 웹 표준 기술입니다.</dd>
<dt>CSS</dt>
<dd>CSS란 Cascading Style Sheet의 약자로, 마크업 요소에 스타일을 적용하기 위해 W3C에서 고안해 낸 언어입니다. 웹 표준에 대한 관심이 높아지면서 CSS를 이용한 디자인도 함께 주목을 받고 있습니다. 그 이유는 CSS가 기존 디자인 방식의 체질 개선을 가능하게 해 줄 수 있기 때문입니다. 특히 CSS는 table 방식의 레이아웃으로 인해 마크업 코드가 복잡해진 표현에 관련된 요소(b 요소와 i 요소 등)의 사용을 제한할 수 있습니다.</dd>
</dl>
</section>
</body>
</html>
```

### figure, figcaption 요소 
- figure 요소는 HTML5에 새롭게 추가된 요소로, 이미지, 비디오 오디오 등을 포함할 수 있으며 figure 요소에 포함된 콘텐츠에 캡션을 제공하고자 할 때 사용할 수 있습니다. 
- 이때 figcaption 요소는 figure 요소의 필수 요소는 아닙니다.

### form
- 폼의 범위를 정의할 때는 form 요소를 사용합니다. 
- form 요소의 속성으로는 폼의 내용을 처리하는 서버의 URI를 지정하기 위한 action과 폼의 내용을 처리하는 방법인 method가 있고, 사용 형식은 다음과 같습니다.

- action=“URI” : 폼의 내용을 처리하기 위한 서버의 URI(필수)
- method=“GET 또는 POST”
  - GET : action 속성에 지정한 URI에 폼의 내용을 추가하여 서버에 송신하는 방식으로, form을 지정할 때 기본값으로 설정되어 있습니다. GET 방식은 데이터의 길이가 짧을 경우에 적합하며, POST 방식보다 처리 속도가 약간 빠릅니다. GET 방식의 경우 웹 브라우저의 주소 창에 변수값이 노출되기 때문에 보안상 중요한 데이터라면 POST 방식을 사용하는 것이 바람직합니다.
  - POST : 폼의 본문으로 송신하는 방식으로, 데이터의 길이가 긴 경우에 적합하며, 보안 문제 때문에 데이터값을 노출시킬 수 없을 때 사용합니다.
- accept–charset : 폼 전송에 사용할 문자 인코딩 지정(문자 인코딩 값을 둘 이상 지정 가능하며, 이때 문자 인코딩 값을 공백으로 구분하여 선언)
- autocomplete : 자동 완성 기능의 활성화 여부(on, off 값 지정)
- enctype : 폼 전송 데이터의 MIME 타입
- name–폼의 이름
- novalidate–폼 전송 시 입력 데이터 값의 검증을 수행하지 않음(논리 속성)
- target–폼의 전송 값을 처리할 대상

```html
<!DOCTYPE html>
<html lang="ko">
<head>
 <meta charset="utf-8">
 <title>form 요소</title>
</head>
<body>
 <section>
  <h1>폼 삽입</h1>
  <form action="login.php" method="post" name="memberLogin"
        accept-charset="utf-8" autocomplete="on"
        target="_blank" novalidate="novalidate">
  <p>
   <label for="userId">아이디</label>
   <input type="text" id="userId">
  </p>
  <p>
   <label for="userPw">비밀번호</label> 
   <input type="password" id="userPw">
  </p>
  <p><input type="submit" value="로그인"></p>
  </form>
 </section> 
</body>
</html>
```

### fieldset, legend
- 여러 개의 폼 요소를 그룹화하여 좀 더 구조적으로 만들려면 fieldset 요소를 사용해야 합니다. 
- fieldset 요소를 이용하면 fieldset의 주위에 border가 생성되어 해당 그룹 안에 있는 콘텐츠가 폼과 관련된 서식임을 쉽게 구별할 수 있습니다. 
- 또 legned 요소로 fieldset에 포함 된 폼 서식이 어떠한 성격인지 제목 형식으로 알려 줄 수 있으며, legend 요소는 fieldset 요소의 자식 요소로 한 번만 지정할 수 있습니다.
- 다만 legend 요소의 경우 크로스 브라우징 관점에서 웹 브라우저 스타일이 다르게 적용되어 있으므로 CSS를 이용한 디자인 작업에는 어려움이 있을 수 있습니다.

```html
<form action="member.asp" method="POST">
 <fieldset>
  <legend>폼 요소의 제목</legend>
   … 생략 …
 </fieldset>
</form>
```

HTML5에서 fieldset 요소에 새롭게 추가된 속성인 form 속성을 사용하면 form 요소 안에 반드시 fieldset 요소를 포함시키지 않아도 해당 form 요소와 연결할 수 있습니다. 

연결하는 방법은 form 요소에는 id 값을 할당하고, fieldset 요소에는 속성으로 form을 지정하며, 해당 속성 값으로 연결하고자 하는 form 요소의 id 값을 지정하면 됩니다.

```html
<form id="memberLogin" action="login.php" method="post" 
      name="memberLogin" accept-charset="utf-8" target="_blank" 
      autocomplete="on" novalidate="novalidate">
</form>
<fieldset form="memberLogin">
 <legend>회원 로그인</legend>
 <p>
  <label for="userId">아이디</label>
  <input type="text" id="userId">
 </p>
 <p>
  <label for="userPw">비밀번호</label>
  <input type="password" id="userPw">
  </p>
 <p><input type="submit" value="로그인"></p>
</fieldset>
```
