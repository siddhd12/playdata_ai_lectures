```python
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup
import time
```

```python
# step1. 아이디, 비밀번호 설정
my_id = "siddhd12@nate.com"
my_pw = "tifhs12!@"

# step2. 크롬 웹드라이버 실행
service = Service(excutable_path = "chromedriver.exe") 
browser = webdriver.Chrome(service = service)

# step3. 인스타그램 로그인 함수 정의
def login(my_id, my_pw):
    # 로그인 페이지로 이동
    browser.get("https://www.instagram.com/")
    time.sleep(1)
    
    # box 및 button 정의
    input_box = browser.find_elements(By.CSS_SELECTOR, "#loginForm > div > div > div > label > input")  # index 0: id, index 1: pw
    btn_login = browser.find_elements(By.CSS_SELECTOR, "#loginForm > div > div > button")
    # print(input_box)
    # print(enter_box)
    time.sleep(1)

    # id 입력
    input_box[0].send_keys(my_id)

    # pw 입력
    input_box[1].send_keys(my_pw)

    # 로그인하기 enter
    btn_login[0].click()
    time.sleep(5)

    # 알림 설정 팝업창 제거 ("나중에 하기 버튼 클릭")
    wait = WebDriverWait(browser, 10)
    btn_later_do_it_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "button._a9--._a9_1")))
    # print(btn_later_do_it_box)
    btn_later_do_it_box.click()
    time.sleep(5)
def insta_hashtag(text):
    url="https://www.instagram.com/explore/tags/"+text
    return url
# step4. 로그인 함수 실행
login(my_id, my_pw)
# btn_click_explorer= browser.find_element(By.CLASS_NAME,'_ab6-')
# btn_click_explorer.click()
word= input()
url=insta_hashtag(word)
browser.get(url)
html = browser.page_source
page = BeautifulSoup(html, "lxml")
```
