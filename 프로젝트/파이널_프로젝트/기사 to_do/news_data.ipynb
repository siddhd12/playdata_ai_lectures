```python
from glob import glob
import json
def get_news_by_sentence(news_folder_path, start_idx, end_idx):
    file_path_list = glob(news_folder_path + "*.json")
    file_path_list = file_path_list[start_idx : end_idx]
    passage_values = []
    for file_path in file_path_list:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            passage = data['Meta(Refine)']['passage'].split('\n')
            stripped = [sen.strip() for sen in passage]
        passage_values.extend(stripped)
    return passage_values, file_path_list
```

```python
file_path = 'C:/Users/Playdata/Desktop/project/성경문체연습/새 텍스트 문서 (2).txt'
num=1
formal=[]
informal=[]
with open(file_path, "r",encoding='utf-8') as file:
    while True:
        content = file.readline()
        if num%2==0:
            informal.append(content.strip())
            num+=1
        elif num%2!=0:
            formal.append(content.strip())
            num+=1
        if content=="":
            break
```

```python
import csv
csv_file_path = "50_60_data.csv"

# CSV 파일에 리스트 저장
with open(csv_file_path, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["formal", "informal"])  # 헤더 행 쓰기
    for formal, informal in zip(formal, informal):
        writer.writerow([formal, informal])  # 데이터 행 쓰기
```
