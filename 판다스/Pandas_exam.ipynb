# Groupby와 집계
## forestfire.csv 파일
- 포르투칼 북서부 산림 산불관련 데이터 셋 [UCI Machine learning repository](https://archive.ics.uci.edu/ml/datasets/forest+fires)    


1. forestfire.csv 파일을 DataFrame으로 읽기
1. 어떤 컬럼들이 범주형 컬럼인지 확인.
1. 월(month)과 요일(day) 의 고유한(unique) 값들과 각각의 행수 조회
1. 요일(day) 별 wind의 평균을 조회
1. 요일(day) 별 wind, rain의 평균을 조회
1. 요일(day) 별 wind, rain의 평균과 합계과 개수 조회
1. 요일(day) 별로 wind는 평균을 rain은 합계와 최대값을 조회
1. 월(month) 그리고 요일별(day) temp의 최대값과 최소값을 조회
1. X좌표(X), Y좌표(Y)별 temp의 평균울 pivot table로 조회


```python
# 1. forestfire.csv 파일을 DataFrame으로 읽기
forest = pd.read_csv("data/forestfire.csv")
forest.info()
```

```python
# 2. 어떤 컬럼들이 범주형 컬럼인지 확인.
# forest["X"].dtype()
# forest["Y"].dtype()
col=forest.columns
type(col)
for x in col:
    print(forest[x].dtype)
```

```python
# 3. 월(month)과 요일(day) 의 고유한(unique) 값들과 각각의 행수 조회
# 범주형 컬럼의 고유값(class)의 확인
forest['month'].unique(),forest['month'].nunique()
```

```python
forest['day'].unique()
```

```python
forest['day'].nunique() # 행수 조회
```

```python
# 4. 요일(day) 별 wind의 평균을 조회
forest.groupby("day")["wind"].mean().to_frame()
```

```python
# 5. 요일(day) 별 wind, rain의 평균을 조회
forest.groupby("day")[["wind","rain"]].mean()
```

```python
# 6. 요일(day) 별 wind, rain의 평균과 합계과 개수 조회
forest.groupby("day")[["wind","rain"]].agg(["mean","sum","count"])
```

```python
# 7. 요일(day) 별로 wind는 평균을 rain은 합계와 최대값을 조회
forest.groupby("day").agg({"wind":"mean","rain":["sum","max"]})
```

```python
# 8. 월(month) 그리고 요일별(day) temp의 최대값과 최소값을 조회
forest.groupby(["month","day"])["temp"].agg(["max","min"])
```

```python
# 9. X좌표(X), Y좌표(Y)별 temp의 평균울 pivot table로 조회
forest.pivot_table(index="X" ,columns="Y",values="temp",aggfunc="mean")
```
